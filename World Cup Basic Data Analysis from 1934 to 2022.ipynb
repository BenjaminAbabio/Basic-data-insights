```python
data4 = data2.sort_values(by=['Total_Attendance','Average_Attendance'], ascending =[True, False])
```

```python
data4.plot(kind='bar', x='Year', y= 'Total_Attendance', figsize = (10, 5), title ='TOTAL ATTENDANCE IN EVERY 4YEARS OF THE WORLD CUP')
```

```python
data4.plot(kind='bar', x='Year', y= 'Average_Attendance', figsize = (10, 5),title ='AVERAGE ATTENDANCE IN EVERY 4YEARS OF THE WORLD CUP')
```

```python
data4.plot(kind='bar', x='Hosts', y= 'Total_Attendance', figsize = (10, 5),title ='TOTAL ATTENDANCE IN EACH HOST COUNTRY SINCE 1934')
```

```python
data4.plot(kind='bar', x='Venue', y= 'Total_Attendance', figsize = (10, 5), title ='TOTAL ATTENDANCE IN VANUES ACROSS ALL THE HOST COUNTRIES') 
```
